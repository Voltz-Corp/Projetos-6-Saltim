## Instruções

- [ ] CC - Entrega de um notebook contendo o carregamento dos dados, análise exploratória inicial e as primeiras etapas de tratamento e transformação dos dados (ex: limpeza, normalização, encoding, eliminação de outliers). A estrutura deve seguir práticas recomendadas de reprodutibilidade e modularização do pipeline

- [ ] CC - Treinamento de um modelo inicial simples utilizando uma abordagem adequada ao tipo de problema definido, como o KNN, regressão linear, árvore de decisão, etc. A entrega deve incluir a aplicação de métricas de desempenho sobre um conjunto de validação e a apresentação dos primeiros resultados

- [ ] CC - Execução de ao menos dois modelos de algoritmos distintos, com objetivo de comparar desempenho com base nas métricas estabelecidas. A entrega deve incluir visualizações que auxiliem a análise comparativa: matriz de confusão, curvas ROC/AUC, gráficos de erro etc.

- [ ] CC - Aplicação de técnicas de ajuste de hiperparâmetros - Grid Search, Random Search, ou Otimização Bayesiana com Optuna - para aprimoramento do desempenho do modelo selecionado. A entrega deve conter a configuração dos parâmetros avaliados e os resultados obtidos

- [ ] CC - Avaliação do modelo por meio de validação cruzada (k-fold ou stratified k-fold), com análise da variância nos resultados e identificação de eventuais problemas de overfitting ou underfitting. Recomenda-se incluir visualizações de curvas de aprendizado ou análise de erro

- [ ] CC - Entrega de uma função de inferência ou estrutura mínima de integração do modelo treinado ao produto da equipe. Pode consistir em um endpoint funcional

- [ ] CC - Execução do pipeline completo com os dados atualizados, utilizando o modelo final ajustado e validadas suas métricas de desempenho. A entrega deve incluir as métricas analisadas e evidências da estabilidade do modelo
